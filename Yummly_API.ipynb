```python
def yummly_get_recipe(app_id, app_key, recipe_list):
    recipes = []
    for recipe in recipe_list:
        url = (f'http://api.yummly.com/v1/api/recipe/{recipe}?_app_id={app_id}&_app_key={app_key}')
        r = requests.get(url)
        response = r.json()
        recipes.append(response)
        
    return recipes
```

```python
def get_recipe_info(recipes, recipe_number):
    try:
        course = recipes[recipe_number]['attributes']['course']
    except:
        course = 'N/A'
    flavors = recipes[recipe_number]['flavors']
    if flavors == {}:
        flavors = 'N/A'
    name = recipes[recipe_number]['name']
    number_of_servings = recipes[recipe_number]['numberOfServings']
    nutrition = recipes[recipe_number]['nutritionEstimates']
    nutrition_fat_kcal = [[nutrition[i]['description'], nutrition[i]['attribute'], nutrition[i]['value'], nutrition[i]['unit']['abbreviation']] \
                          for i, x in enumerate(nutrition) if (nutrition[i]['attribute'] == 'FAT_KCAL')]
    nutrition_sugar_g = [[nutrition[i]['description'], nutrition[i]['attribute'], nutrition[i]['value'], nutrition[i]['unit']['abbreviation']] \
                          for i, x in enumerate(nutrition) if (nutrition[i]['attribute'] == 'SUGAR')]
    nutrition_energykj_kcal = [[nutrition[i]['description'], nutrition[i]['attribute'], nutrition[i]['value'], nutrition[i]['unit']['abbreviation']] \
                          for i, x in enumerate(nutrition) if (nutrition[i]['attribute'] == 'ENERC_KJ')]
    nutrition_fat_g = [[nutrition[i]['description'], nutrition[i]['attribute'], nutrition[i]['value'], nutrition[i]['unit']['abbreviation']] \
                          for i, x in enumerate(nutrition) if (nutrition[i]['attribute'] == 'FAT')]
    nutrition_protein_g = [[nutrition[i]['description'], nutrition[i]['attribute'], nutrition[i]['value'], nutrition[i]['unit']['abbreviation']] \
                          for i, x in enumerate(nutrition) if (nutrition[i]['attribute'] == 'PROCNT')]
    nutrition_carb_g = [[nutrition[i]['description'], nutrition[i]['attribute'], nutrition[i]['value'], nutrition[i]['unit']['abbreviation']] \
                          for i, x in enumerate(nutrition) if (nutrition[i]['attribute'] == 'CHOCDF')]
    nutrition_cholesterol_g = [[nutrition[i]['description'], nutrition[i]['attribute'], nutrition[i]['value'], nutrition[i]['unit']['abbreviation']] \
                          for i, x in enumerate(nutrition) if (nutrition[i]['attribute'] == 'CHOLE')]
    nutrition_energy_kcal = [[nutrition[i]['description'], nutrition[i]['attribute'], nutrition[i]['value'], nutrition[i]['unit']['abbreviation']] \
                          for i, x in enumerate(nutrition) if (nutrition[i]['attribute'] == 'ENERC_KCAL')]
    
    # print(f'Recipe name: {name} \n Course type: {course} \n Flavor profiles: {flavors} \n No. of servings: {number_of_servings} \n Nutritional info: {nutrition_list}')
    
    return [name, course, flavors, number_of_servings, nutrition_fat_kcal, nutrition_sugar_g, nutrition_energykj_kcal, nutrition_fat_g, nutrition_protein_g, nutrition_carb_g, nutrition_cholesterol_g, nutrition_energy_kcal]
```

```python
pizza_recipes = []

for i in range(len(recipes)):
    recipe_info = get_recipe_info(recipes, i)
    pizza_recipes.append(recipe_info)
```

```python
columns = ['name', 'course', 'flavors', 'number_of_servings', 'fat_kcal', 'sugar_g', 'energykj_kcal', 'fat_g', 'protein_g', 'carbs_g', 'cholesterol_g', 'energy_kcal']

first_100_pizza_df = pd.DataFrame(pizza_recipes, columns = columns)
```

```python
first_100_pizza_df
```

```python
def get_nutritional_value(row):
    if row != []:
        value = row[0][2]
    else:
        value = np.nan
    return value
```

```python
first_100_pizza_df.fat_kcal = first_100_pizza_df.fat_kcal.apply(get_nutritional_value)
first_100_pizza_df.sugar_g = first_100_pizza_df.sugar_g.apply(get_nutritional_value)
first_100_pizza_df.energykj_kcal = first_100_pizza_df.energykj_kcal.apply(get_nutritional_value)
first_100_pizza_df.fat_g = first_100_pizza_df.fat_g.apply(get_nutritional_value)
first_100_pizza_df.protein_g = first_100_pizza_df.protein_g.apply(get_nutritional_value)
first_100_pizza_df.carbs_g = first_100_pizza_df.carbs_g.apply(get_nutritional_value)
first_100_pizza_df.cholesterol_g = first_100_pizza_df.cholesterol_g.apply(get_nutritional_value)
first_100_pizza_df.energy_kcal = first_100_pizza_df.energy_kcal.apply(get_nutritional_value)
```
